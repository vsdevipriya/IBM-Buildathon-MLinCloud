```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/ObesityDataSet.csv')
```

```python
df['Gender']= df['Gender'].replace({'Male':0, 'Female':1})
df['FAVC']=df['FAVC'].replace({'yes':1,'no':0})
df['family_history_with_overweight']= df['family_history_with_overweight'].replace({'yes':1, 'no':0})
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['CAEC']= le.fit_transform(df['CAEC'])
df['SMOKE']=le.fit_transform(df['SMOKE'])
df['SCC']=le.fit_transform(df['SCC'])
df['CALC']=le.fit_transform(df['CALC'])
pd.get_dummies(df['MTRANS'])
df['MTRANS']=le.fit_transform(df['MTRANS'])
df['NObeyesdad']=le.fit_transform(df['NObeyesdad'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('NObeyesdad',axis=1)
y=df['NObeyesdad']
```

```python
x
```

```python
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=12)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
xtrain
```

```python
dt= DecisionTreeClassifier()
rf=RandomForestClassifier()
```

```python
dt.fit(xtrain,ytrain)
rf.fit(xtrain,ytrain)
```

```python
ypred=dt.predict(xtest)
yrandompredict=rf.predict(xtest)
```

```python
ypred
```

```python
yrandompredict
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(ytest,ypred))
```

```python
print(classification_report(ytest,yrandompredict))
```

```python
confusion_matrix(ytest,ypred)
```

```python
confusion_matrix(ytest,yrandompredict)
```
